# KNN - Predict whether a person will have diabetes or not

# Import all dependencies

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix, accuracy_score
```

# Load diabetes.csv

```python
df = pd.read_csv('diabetes.csv')
```

## Examine the dataset

```python
df.head()
```

## Data processing

```python
# Replace zeroes
zero_not_accepted = ['Glucose', 'BloodPressure', 'SkinThickness', 'BMI', 'Insulin']
```

### 1. Compute the mean 
### 2. replace the above columns' value with the mean.


```python
for column in zero_not_accepted:
    mean_value = df[column].replace(0, pd.NA).astype(float).mean()#去掉0
    df[column] = df[column].replace(0, mean_value)
```

### Split the dataset with Test and Train with test_size=0.2

```python
x = df.drop('Outcome', axis=1)
y = df['Outcome']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

### Train the model

```python
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
knn = KNeighborsClassifier(n_neighbors=5)  
knn.fit(x_train, y_train)
```

### Predict the model with test data

```python
y_pred = knn.predict(x_test)
print(y_pred)
```

### Evaluate the model

```python
print("accuracy:", accuracy_score(y_test, y_pred))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))
```
